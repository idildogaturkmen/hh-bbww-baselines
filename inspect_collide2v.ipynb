Then match those two b quarks to AK4 jets.

```python
import numpy as np
import pandas as pd

def delta_phi(phi1, phi2):
    ''' computes the angular difference in phi, but correctly handles the wraparound at ±π.
    '''
    return np.arctan2(np.sin(phi1 - phi2), np.cos(phi1 - phi2))

def delta_r(eta1, phi1, eta2, phi2):
    '''ΔR = sqrt((Δη)^2 + (Δφ)^2)'''
    return np.sqrt((eta1 - eta2)**2 + delta_phi(phi1, phi2)**2)

# delta R to match a generator-level b quark to the nearest reconstructed AK4 jet.

jets = pd.DataFrame({
    "idx": range(len(row["FullReco_JetAK4_PT"])),
    "pt": row["FullReco_JetAK4_PT"],
    "eta": row["FullReco_JetAK4_Eta"],
    "phi": row["FullReco_JetAK4_Phi"],
    "mass": row["FullReco_JetAK4_Mass"],
    "btag": row["FullReco_JetAK4_BTag"],
    "btag_phys": row["FullReco_JetAK4_BTagPhys"],
})

matches = []

for _, b in b_from_higgs.iterrows():
    drs = delta_r(jets["eta"], jets["phi"], b["eta"], b["phi"])
    best = drs.idxmin()

    matches.append({
        "b_gen_idx": b["idx"],
        "b_pid": b["pid"],
        "b_pt": b["pt"],
        "b_eta": b["eta"],
        "b_phi": b["phi"],
        "matched_jet_idx": jets.loc[best, "idx"],
        "jet_pt": jets.loc[best, "pt"],
        "jet_eta": jets.loc[best, "eta"],
        "jet_phi": jets.loc[best, "phi"],
        "jet_btag": jets.loc[best, "btag"],
        "jet_btag_phys": jets.loc[best, "btag_phys"],
        "deltaR": drs.loc[best],
    })

matches = pd.DataFrame(matches)
matches
```

Now, repeat this logic for one event in a function

```python
def get_hbb_label(row, dr_cut=0.4):
    gen = pd.DataFrame({
        "idx": range(len(row["FullReco_GenPart_PT"])),
        "pt": row["FullReco_GenPart_PT"],
        "eta": row["FullReco_GenPart_Eta"],
        "phi": row["FullReco_GenPart_Phi"],
        "pid": row["FullReco_GenPart_PID"],
        "m1": row["FullReco_GenPart_M1"],
        "m2": row["FullReco_GenPart_M2"],
        "d1": row["FullReco_GenPart_D1"],
        "d2": row["FullReco_GenPart_D2"],
        "status": row["FullReco_GenPart_Status"],
    })

    jets = pd.DataFrame({
        "idx": range(len(row["FullReco_JetAK4_PT"])),
        "pt": row["FullReco_JetAK4_PT"],
        "eta": row["FullReco_JetAK4_Eta"],
        "phi": row["FullReco_JetAK4_Phi"],
        "mass": row["FullReco_JetAK4_Mass"],
        "btag": row["FullReco_JetAK4_BTag"],
        "btag_phys": row["FullReco_JetAK4_BTagPhys"],
    })

    if len(jets) < 2: # Require at least 2 jets for H -> bb
        return None

    # finds the Higgs bosons at the generator stage that seem to correspond to the useful final hard-process Higgs copies
    higgs_final = gen[(gen["pid"].abs() == 25) & (gen["status"] == 62)] 
    if len(higgs_final) == 0:
        return None

    # Get Higgs daughter branch IDs to find b quarks from Higgs decay
    higgs_branch_ids = set(higgs_final["d1"].tolist() + higgs_final["d2"].tolist())

    # Find b quarks from those Higgs branches
    b_from_higgs = gen[
        (gen["pid"].abs() == 5)
        & (gen["m1"].isin(higgs_branch_ids))
    ]

    if len(b_from_higgs) != 2:
        return None

    matched_jets = []

     # Match each b quark to the nearest jet and check if it's within the delta R cut
    for _, b in b_from_higgs.iterrows():
        drs = delta_r(jets["eta"], jets["phi"], b["eta"], b["phi"])
        best = drs.idxmin()

        # If the closest jet is too far away, then this event isn't a good H->bb candidate
        if drs.loc[best] > dr_cut:
            return None

        matched_jets.append(int(jets.loc[best, "idx"]))

    # rejects cases where both b quarks match to the same jet.
    if len(set(matched_jets)) != 2:
        return None

    return {
        "hbb_jets": matched_jets,
        "n_jets": len(jets),
        "jets": jets,
    }
```

Run for 100 events

```python
n_seen = 0
n_usable = 0
examples = []

for row in ds:
    n_seen += 1
    label = get_hbb_label(row)

    if label is not None:
        n_usable += 1
        examples.append(label)

    if n_seen >= 100:
        break

print("seen:", n_seen)
print("usable:", n_usable)
print("usable fraction:", n_usable / n_seen)
```

#### mass-pairing baseline
Pick the jet pair whose invariant mass is closest to 125 GeV.

```python
import itertools
import numpy as np

def fourvec(pt, eta, phi, mass):
    px = pt * np.cos(phi)
    py = pt * np.sin(phi)
    pz = pt * np.sinh(eta)
    E = np.sqrt(px**2 + py**2 + pz**2 + mass**2)
    return E, px, py, pz

def dijet_mass(j1, j2):
    E1, px1, py1, pz1 = fourvec(j1.pt, j1.eta, j1.phi, j1.mass)
    E2, px2, py2, pz2 = fourvec(j2.pt, j2.eta, j2.phi, j2.mass)
    m2 = (E1+E2)**2 - (px1+px2)**2 - (py1+py2)**2 - (pz1+pz2)**2
    return np.sqrt(max(m2, 0))

def best_mass_pair(jets, target=125):
    best_pair = None
    best_score = float("inf")
    for i, j in itertools.combinations(jets.index, 2):
        m = dijet_mass(jets.loc[i], jets.loc[j])
        score = abs(m - target)
        if score < best_score:
            best_score = score
            best_pair = sorted([int(jets.loc[i, "idx"]), int(jets.loc[j, "idx"])])
    return best_pair
```

Compare b-tag to mass baseline

```python
results = {"btag": 0, "mass": 0}
total = 0

for label in examples:
    jets = label["jets"]
    truth = sorted(label["hbb_jets"])

    pred_btag = sorted(jets.sort_values("btag", ascending=False)["idx"].head(2).tolist())
    pred_mass = best_mass_pair(jets)

    results["btag"] += pred_btag == truth
    results["mass"] += pred_mass == truth
    total += 1

print("total:", total)
print("btag accuracy:", results["btag"] / total)
print("mass accuracy:", results["mass"] / total)
```

## Try: ΔR pairing
smallest ΔR between jets
because H → bb often produces relatively collimated jets

```python
def best_dr_pair(jets):
    best_pair = None
    best_dr = float("inf")

    for i in range(len(jets)):
        for j in range(i+1, len(jets)):
            dr = delta_r(
                jets.loc[i, "eta"], jets.loc[i, "phi"],
                jets.loc[j, "eta"], jets.loc[j, "phi"]
            )
            if dr < best_dr:
                best_dr = dr
                best_pair = sorted([jets.loc[i, "idx"], jets.loc[j, "idx"]])

    return best_pair
```

```python
# evaluate this DR-based pairing method as well
results["dr"] = 0
total = 0

for label in examples:
    jets = label["jets"]
    truth = sorted(label["hbb_jets"])

    pred_dr = best_dr_pair(jets)

    results["dr"] += pred_dr == truth
    total += 1

print("DR pairing accuracy:", results["dr"] / total)
```

combined baseline next:

```python
def best_btag_mass_pair(jets, target=125, alpha=50):
    best_pair = None
    best_score = float("inf")

    for i in range(len(jets)):
        for j in range(i + 1, len(jets)):
            m = dijet_mass(jets.loc[i], jets.loc[j])
            bsum = jets.loc[i, "btag"] + jets.loc[j, "btag"]
            score = abs(m - target) - alpha * bsum

            if score < best_score:
                best_score = score
                best_pair = sorted([int(jets.loc[i, "idx"]), int(jets.loc[j, "idx"])])

    return best_pair
```

```python
correct = 0
total = 0

for label in examples:
    jets = label["jets"]
    truth = sorted(label["hbb_jets"])

    pred = best_btag_mass_pair(jets)
    correct += pred == truth
    total += 1

print("btag+mass accuracy:", correct / total)
```

#### Question: Does pT help fix pairing?

```python
def top2_pt(jets):
    return sorted(jets.sort_values("pt", ascending=False)["idx"].head(2).tolist())
```

```python
correct = 0
total = 0

for label in examples:
    jets = label["jets"]
    truth = sorted(label["hbb_jets"])

    pred = top2_pt(jets)
    correct += pred == truth
    total += 1

print("pt accuracy:", correct / total)
```

Picking the two highest-pT jets gives the correct H→bb pair only ~14% of the time. So pT alone is a weak reconstruction rule. The Higgs b-jets are not simply the two hardest jets in the event.

##### Restrict candidate jets

```python
def top4_btag_then_mass(jets):
    jets_top = jets.sort_values("btag", ascending=False).head(4)

    return best_mass_pair(jets_top)
```
